# Lectura de datos

```python
def abrirArchivos(ruta, letra, columnas):
    """ En esta función se abren los archivos con las columnas seleccionadas 
        ruta: es la ruta del archivo
        letra: es la identifiación del tipo de dato -> P=Precipitacion, T=Temperatura
        return df_datos: retorna los datos seleccionados
    """
    if letra == "P":
        df_datos = pd.read_csv(ruta, usecols = columnas)
        df_datos = df_datos.fillna(-1)
        #df_datos.fecha = pd.to_datetime(df_datos.fecha)

    else:
        df_datos = pd.read_csv(ruta, usecols = columnas)
        df_datos = df_datos.fillna(-99)
        #df_datos.fecha = pd.to_datetime(df_datos.fecha, usecols = columnas)

    return df_datos

#--------------------------Lectura datos de precipitación-------------------------------------
rutaP = 'C:/Users/Bender/Desktop/Mapas_ERA/Anexos_Final/Anexo A series IDEAM bruta/precipitacion_sup.csv'
precpGet = ['21210110', '21210120', '21210130']
df_precipSup = abrirArchivos(rutaP, "P", precpGet)

#--------------------------Lectura datos de temperatura---------------------------------------
rutaT = 'C:/Users/Bender/Desktop/Mapas_ERA/Anexos_Final/Anexo A series IDEAM bruta/temperatura_sup.csv'
tempGet = ['22050100', '23010020', '23010080']
df_tempSup = abrirArchivos(rutaT, "T", tempGet)
```

# Generar archivo tetis de evento de datos en superficie

```python
#--------------------------------Generar archivo de evento txt-----------------------------------------
def archivoEvento(df_precip_sup, df_tempSup, df_ubi_info, precpGet, tempGet):
    #---------ENCABEZADO-----------------------------------------------
    Tetis_file = open("Tetis_file.txt","w+")
    Tetis_file.write("G               "+str(len(df_precip_sup))+"         1440\n")
    Tetis_file.write("*         dd-mm-aaaa  hh:mm\n")
    Tetis_file.write("F           01-01-1950      00:00\n")
    
    #---------------------COLOCAR LOS DATOS DE UBICACION DE LOS PUNTOS O ESTACIONES------------------------------------------------------
    #Precipitacion
    for i in precpGet:
        Tetis_file.write('P         "'+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == i,'CODIGO'].values[0])   +'" '+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == i,'latitud'].values[0])   +' '+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == i,'longitud'].values[0])   +'      '+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == i,'altitud'].values[0])   +'\n')  
    #Temperatura
    for j in tempGet:
        Tetis_file.write('T         "'+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == j,'CODIGO'].values[0])   +'" '+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == j,'latitud'].values[0])   +' '+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == j,'longitud'].values[0])   +'      '+str(df_ub_precipSup.loc[df_ub_precipSup['CODIGO'] == j,'altitud'].values[0])   +'\n')
    
    #-----------------------------INGRESAR LOS DATOS-------------------------------------------------------
    df_total = pd.concat([df_precipSup, df_tempSup], axis = 1)
    Tetis_file.write("*dt(dia)\n")
    Tetis_file.write('* desde:  01/01/1950  hasta:  31/12/2021\n')
    Tetis_file.write('*')
    Tetis_file.write(df_total[df_total.columns].to_string(col_space=7))
    #Tetis_file.write(df_total[precpGet[:]].to_string(col_space=7))

    Tetis_file.close()

    if df_precip_sup == 0:
        print(1)

    else:
        print(0)
    return Tetis_file
```

```python
def calc_evapot(df_tempSup,df_ubi,df_radiacion):
   df_tempSup['CODIGO'] = pd.date_range(start='01/01/1950', periods=len(df_tempSup), freq='D')
   df_tempSup.set_index(['CODIGO'], inplace=True)
   HR = [82.065943, 82.91073, 85.125509, 85.536313, 85.341042, 83.060406, 80.410283, 78.7061, 79.407464, 83.32468, 85.336647, 84.15651]
   meses = ['Ene',	'Feb',	'Mar',	'Abr',	'May',	'Jun',	'Jul',	'Ago',	'Sep',	'Oct',	'Nov',	'Dic']
   K = [0.4, 0.37, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4]
   #Generar las isotermas
   #Agrupar por mes y anio
   df_temp_mes_anio = df_tempSup.groupby(pd.Grouper(freq = 'M')).mean()
   #Calcular la media y dejar solo 12 meses
   df_temp_mes_med = df_temp_mes_anio.groupby([(df_temp_mes_anio.index.month)]).mean().T
   
   df_ubi.set_index('CODIGO', inplace=True)

   df_evapot = pd.merge(df_temp_mes_med, df_ubi, left_index=True, right_index=True)

   df_radiacion = pd.concat([df_radiacion,df_evapot[['latitud']]],ignore_index = False).sort_values(['latitud'], ascending=True)

   df_radiacion = df_radiacion.interpolate()

   df_evapot = pd.merge(df_evapot, df_radiacion, left_index=True, right_index=True)

   for n, j, i, m in zip(range(1,13), K, HR , meses):

      df_evapot[[n]] = (j*(df_evapot[[n]]/(df_evapot[[n]]+15))) * (1+((50-i)/70))

      df_evapot[[str(m)]] = (df_evapot[[str(m)]]+50)

      df_evapot[['new' + str(n)]] = df_evapot[n] * df_evapot[str(m)]







   return df_evapot
```
